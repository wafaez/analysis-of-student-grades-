### Supprimant l'etudiant qui s'abscente dans toutes les modules

```python
def Supprimer_NaN(df):
    df1 = df.dropna(thresh=4)
    df1.reset_index(drop=True, inplace=True)
    return df1

df2 = Supprimer_NaN(df)
df2.shape

```

### Calcul et affichage de la moyenne de toutes les notes

```python

def Moyenne(df):
    df['Moyenne'] = (df.sum(numeric_only=True, axis=1))/10
    return df
new_df=Moyenne(df)
new_df
```

### Valid√© ou Ajourn√©

```python
def resultat(new_df):
    new_df["R√©sultat"]=["ajourn√©" if x<10 else "valid√©" for x in new_df["Moyenne"]]
    return new_df
new_df=resultat(new_df)
new_df
```

### diagramme Camembert

```python
def diag_camembert(new_df):
    colors=[ '#5D7052', '#C18845']
    df2=new_df.R√©sultat.value_counts()
    labels = df2.index
    explode = (0, 0.2)  # on isole seulement la deuxi√®me part (c.a.d 'ajourne')
    fig, ax = plt.subplots()
    ax.pie(df2, labels=labels, explode=explode, 
    autopct='%1.1f%%', shadow=True,colors=colors,startangle=45)
    ax.axis('equal')
    plt.title('R√©sultat', fontsize=14)
    plt.show()
diag_camembert(new_df)
```

### Ajouter Mention 

```python
Mention=[]
for i in new_df.index:
    if 10<=new_df.loc[i,'Moyenne']<12:
        Mention.append("Passable")
    elif 12<=new_df.loc[i,'Moyenne']<14:
        Mention.append("Assez_bien")
    elif 14<=new_df.loc[i,'Moyenne']<16:
        Mention.append("Bien")
    elif 16<=new_df.loc[i,'Moyenne']<18:
        Mention.append("Tres_bien")
    elif 18<=new_df.loc[i,'Moyenne']<20:
        Mention.append("Excelent")
    else :
        Mention.append("____")
new_df['Mention']=Mention
new_df
```

```python
plt.figure(figsize=(15,6))
sns.set_context("talk",font_scale=1)
sns.set_palette("pastel")
ax = sns.countplot(y="Mention", x=None, hue="Genre", data=new_df, order=["Passable","Assez_bien","Bien","Tres_bien","Excelent"])
ax.legend(loc='upper right',frameon=True)
plt.title('Genre vs Mention', fontsize=15, fontweight='bold')
ax.set(xlabel='nombre',ylabel='mention')
plt.show()
```
